```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x     = np.linspace(-1.0, 1.0, 100, endpoint=True)
y     = np.linspace(-1.0, 1.0, 100, endpoint=True)
X, Y  = np.meshgrid(x,y)
```

```python
func  = lambda x, y: x**2 - y**2
```

```python
Z     = func(X,Y)
```

```python
f,a = plt.subplots()
cs  = a.contourf(X,Y,Z)
cb = f.colorbar(cs, ax=a)
```

## Customization

```python
levels = [-0.5, 0.0, 0.5]
colors = ('C0', 'C1')
extend = 'neither'
alpha = 0.9
```

```python
f,a = plt.subplots()
cs  = a.contourf(X,Y,Z, levels=levels, colors=colors, extend=extend, alpha=alpha)
cb = f.colorbar(cs, ax=a)
```
