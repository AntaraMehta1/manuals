```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(-5.0, 5.0, 100, endpoint=True)
y = np.linspace(-5.0, 5.0, 100, endpoint=True)
```

```python
X,Y = np.meshgrid(x,y)
```

```python
def f(x,y):
    f = x**2 - y**2
    return f
```

```python
Z = f(X,Y)
```

```python
f,a = plt.subplots(figsize=(8.0,8.0))
CS = a.contour(X,Y,Z, levels=[-10.0, 0.0, 10.0])
a.clabel(CS, inline=True)
```
